```python
def clamp(n, smallest, largest): 
    return max(smallest, min(n, largest))
```

```python
def getRevenue(capacity=12500,percentInvested=0.0,benjaminBuff=0.0):
    rate = 1.15
    cashPerRound = (325/1.05)*(1+benjaminBuff)
    
    startingCash = capacity*percentInvested
    
    finalCash = (startingCash+cashPerRound)*rate
    
    profit = clamp(finalCash,0,capacity)-startingCash
    
    return round(profit)
```

```python
def make_df(*farms):
    images = []
    upgrades = []
    costs = []
    revenues = []
    interestRates = []
    descriptions = []
    
    for i in range(len(farms)):
        interest = round((farms[i][3]/farms[i][2])*100,1)
        description = farms[i][4] if len(farms[i])>4 else ""
        
        images.append(farms[i][0])
        upgrades.append(farms[i][1])
        costs.append(farms[i][2])
        revenues.append(farms[i][3])
        interestRates.append(f"{interest}%")
        descriptions.append(description)
        
    df = pd.DataFrame({
        'Image':images,
        'Upgrades':upgrades,
        'Cost (hard)':costs,
        'Cash/Round':revenues, 
        'Interest':interestRates,
        'Description':descriptions,
    })
    return HTML(df.to_html(escape=False))
```

```python
class Bank:
    def __init__(self, principal, bank_hack = 1):
        """set bank hack to 1.05 or 1.12 if using benjamin"""
        self.principal = principal
        self.cash = principal
        self.interest_rate = 1.15
        self.cash_limit = 9500
        self.income = 310 
        self.income *= bank_hack
        
    def compound(self, interest=True):
        """increases cash level by adding income then adding interest"""
        
        if interest:
            self.cash += self.income
            self.cash *= self.interest_rate
        else:
            income = self.income*self.interest_rate
            self.cash += income
        
    def estimate_earnings(self, max_out=False):
        earnings = [self.cash]
        self.compound(False)
        while self.cash < self.cash_limit:
            earnings.append(self.cash)
            self.compound()
            
        if max_out:
            earnings.append(self.cash_limit)
            
        return Earnings(earnings)
    
    @property
    def cost(self):
        """assuming 2:3:0 bank"""
        return 1225 + 540 + 650 + 325 + 865 + 3565 + self.principal

            
    def __repr__(self):
        return "${}".format(self.cash)
        
```

```python
class LoanBank(Bank):
    def __init__(self, principal, bank_hack=1):
        super().__init__(principal, bank_hack)
        self.cash_limit = 12500
        
    @property
    def cost(self):
        """assuming 2:4:0 bank"""
        return 1225 + 540 + 650 + 325 + 865 + 3565 + 8100 + self.principal 
        
```

```python
class Earnings:
    def __init__(self, record):
        self.record = record
        
    @property
    def principal(self):
        return self.record[0]
    
    @property
    def revenue(self):
        return self.record[-1]
    
    @property
    def profit(self):
        return self.revenue-self.principal
    
    @property
    def rounds(self):
        return len(self.record) - 1
    
    @property
    def cash_per_round(self):
        return self.profit/self.rounds
    
    @property
    def cost(self):
        """assuming 2:3:0 or 2:4:0 bank"""
        cost = 1225 + 540 + 650 + 325 + 865 + 3565 + self.principal
        if 9500 < self.revenue <= 12500:
            # 2:4:0
            return cost + 8100
        return cost
    
    @property
    def interest(self):
        return (self.cash_per_round/self.cost) * 100
    
    def future_value(self, rounds):
        return (1+(self.interest/100)/(1/self.rounds))**(rounds*(1/self.rounds))
    
    def __gt__(self, other):
        assert isinstance(other, Earnings)
        return self.future_value(44) > other.future_value(44)
        
    def __repr__(self):
        string = ""
        string += f"Avg cash/round: {self.cash_per_round}\n"
        string += f"Avg interest: {self.interest}\n"
        string += f"Rounds until full: {self.rounds} \n\tie buy bank at start of round 3, pull money out at start of round {3 + self.rounds}\n\n"
        string += repr(self.record)
        return string
        
```

```python
# after 44 rounds which bank strategy would make the most cash?
# each strategy starts with 1 bank
# strategy A uses no deposit banks and reinvests all earnings into making more banks
# strategy B fills the banks to half full and reinvests all earnings into half full banks



class Strat:
    def __init__(self, initial_investment):
        self.cash = 100000
        self.rounds = 3
        self.banks = []
        self.initial_investment = initial_investment
        
    def __repr__(self):
        return f"Round {self.rounds}: {len(self.banks)} banks with ${round(self.cash)} left over"

    
    def buy_banks(self):
        bank_cost = b(self.initial_investment).cost
        num_banks = int(self.cash // bank_cost)
        for _ in range(num_banks):
            self.banks.append(b(self.initial_investment))
            self.cash -= bank_cost
            
    def next_revenue(self):
        for bank in self.banks:
            self.cash += bank.profit
        self.rounds += self.banks[0].rounds
        
    def sell_all(self, percent_liquidity = 1):
        for bank in self.banks:
            self.cash += bank.cost*percent_liquidity
        self.banks = []

        
            
        
def b(cash, bank_hack=1):
    return Bank(cash,bank_hack).estimate_earnings()


zero_banks = Strat(0)
half_banks = Strat(4750)
threeQ_banks = Strat(7125)

def investment_strat(zero_banks):
    while zero_banks.rounds < 47:
        zero_banks.buy_banks()
        print(zero_banks)

        zero_banks.next_revenue()
        print(zero_banks)
    
    zero_banks.sell_all()
    print(zero_banks)
    print()

investment_strat(zero_banks)
investment_strat(half_banks)
investment_strat(threeQ_banks)
# zero_banks.buy_banks()
# half_banks.buy_banks()
# print(zero_banks)
# print(half_banks)


# wait 4 more rounds or 11 more rounds
# print total cash
# buy banks again

# sell all banks at the end (whats the sell value?)
```
